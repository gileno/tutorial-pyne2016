```python
%matplotlib inline
import matplotlib.pyplot as pl
import pandas as pd
import statsmodels.api as sm
import seaborn

df = pd.read_csv("dataset/copacabana.csv", sep=";")
```

```python
df.describe()
```

```python
Y = df["VAL_UNIT"]
```

```python
X = df[["Quartos", "Vagas", "DistIpanema", "DistPraia", "DistFavela", "RendaMedia", "Mes", "Idade"]]
```

```python
Y
```

```python
X
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
model = sm.OLS(Y, X)
results = model.fit()
```

```python
results.params
```

```python
predict = model.predict(results.params, X)
```

```python
predict = pd.DataFrame(predict)
```
